
## Ejercicio 1


```matlab
clc; clear all; close all;
Origin = [0,0];
s = 3;
Axis = [-s,s];
R = 15;
lw = 1.5;
P = [1; -2; 1; 1];
figure
for i = 1:R
  hold off
  plot3(Axis, Origin, Origin, 'Linewidth', lw);
  hold on
  plot3(Origin, Axis, Origin, 'Linewidth', lw);
  plot3(Origin, Origin, Axis, 'Linewidth', lw);
  Ry = [cosd(i),0, sind(i), 0;
    0, 1, 0, 0;
    -sind(i), 0, cosd(i), 0;
    0, 0, 0, 1];
  P1 = Ry * P;
  plot3([0,P(1)],[0,P(2)],[0,P(3)], "LineWidth", lw);
  plot3([0,P1(1)],[0,P1(2)],[0,P1(3)], "LineWidth", lw);
  title(['Value of the Norm 2 of P1: ', num2str(norm(P1(1:3)))])
  legend("X Axis", "Y Axis", "Z Axis", "Original Vector", "Final Vector")
  pause(0.5);
end
```

## Ejercicio 2


```matlab
clc; clear all; close all;
Origin = [0,0];
s = 5; %Size of the Axis
Axis = [-s,s];
T = 2; %Transformation
lw = 1.5;
P = [2;3;-1; 1];
for i = 0:0.5:T
    hold off
    plot3(Axis, Origin, Origin, 'Linewidth', lw);
    hold on
    plot3(Origin, Axis, Origin, 'Linewidth', lw);
    plot3(Origin, Origin, Axis, 'Linewidth', lw);
    Tz = eye(4);
    Tz(3,4) = i;
    P1 = Tz * P;
    plot3([0,P(1)],[0,P(2)],[0,P(3)], "LineWidth", lw);
    plot3([0,P1(1)],[0,P1(2)],[0,P1(3)], "LineWidth", lw);
    title(['Value of the Norm 2 of P1: ', num2str(norm(P1(1:3)))])
    legend("X Axis", "Y Axis", "Z Axis", "Original Vector", "Final Vector")
    view([-50,50])
    pause(0.5);
end
```
